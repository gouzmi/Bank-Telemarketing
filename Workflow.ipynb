```python
import numpy as np
import pandas as pd
import sys
import warnings
warnings.filterwarnings('ignore')

from sklearn.base import BaseEstimator
from sklearn.base import TransformerMixin
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

import problem
```

## Get Data

```python
X, y = problem.get_train_data()
X_test, y_test = problem.get_test_data()
```

## Data preprocessing

```python
from sklearn.impute import KNNImputer

# convert categorical columns to categorical type
cat_cols = [column_name for column_name in X.columns if str(
    X[column_name].dtype) == 'object']
num_cols = [
    column_name for column_name in X.columns if column_name not in cat_cols]
X[cat_cols] = X[cat_cols].astype('category')
```

```python
# one hot encode to be able to use KNNImputation
X = pd.get_dummies(X, dummy_na=True)
for col in cat_cols:
    X.loc[X[col + "_nan"] == 1, X.columns.str.startswith(col)] = np.nan
    del X[col + "_nan"]
```

```python
# impute values
imputer = KNNImputer(n_neighbors=5)
X = pd.DataFrame(imputer.fit_transform(X.values),columns = X.columns)

# revert dummification
for col in cat_cols:
    X[col] = X.loc[:, X.columns.str.startswith(col)].idxmax(axis=1).str.replace(col + "_", '')
    X = X.loc[:,~X.columns.str.startswith(col + "_")] 

# reset categorical column types
X[cat_cols] = X[cat_cols].astype('category')
```
