````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

pd.options.display.max_columns = None
````

Let's load the data and take a first look at it.

````python
df = pd.read_csv('data/bank-full.csv', sep=";")
````

````python
df.head()
````

Here is the official description of the attributes of our dataset.


#### Bank client data:

- **1 - age (numeric)**
- **2 - job (categorical)** : type of job ('admin.', 'blue-collar', 'entrepreneur', 'housemaid', 'management', 'retired', 'self-employed', 'services', 'student', 'technician', 'unemployed', 'unknown')
- **3 - marital (categorical)** : marital status ('divorced', 'married', 'single', 'unknown'; note: 'divorced' means divorced or widowed)
- **4 - education (categorical)** : 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'illiterate', 'professional.course', 'university.degree', 'unknown')
- **5 - default (categorical)** : has credit in default? ('no', 'yes', 'unknown')
- **6 - housing (categorical)** : has housing loan? ('no', 'yes', 'unknown')
- **7 - loan (categorical)** : has personal loan? ('no', 'yes', 'unknown')

#### Related with the last contact of the current campaign:

- **8 - contact (categorical)** : contact communication type ('cellular', 'telephone')
- **9 - month (categorical)**: last contact month of year ('jan', 'feb', 'mar', ..., 'nov', 'dec')
- **10 - day_of_week (categorical)** : last contact day of the week ('mon', 'tue', 'wed', 'thu', 'fri')
- **11 - duration (numeric)** : last contact duration, in seconds. Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

#### Other attributes:
- **12 - campaign (numeric, includes last contact)** : number of contacts performed during this campaign and for this client
- **13 - pdays (numeric; 999 means client was not previously contacted)** : number of days that passed by after the client was last contacted from a previous campaign 
- **14 - previous (numeric)** : number of contacts performed before this campaign and for this client
- **15 - poutcome (categorical)** : outcome of the previous marketing campaign ('failure', 'nonexistent', 'success')

#### Social and economic context attributes

- **16 - emp.var.rate (numeric)** : employment variation rate - quarterly indicator
- **17 - cons.price.idx (numeric)** : consumer price index - monthly indicator
- **18 - cons.conf.idx (numeric)** : consumer confidence index - monthly indicator
- **19 - euribor3m (numeric)** : euribor 3 month rate - daily indicator
- **20 - nr.employed (numeric)** : number of employees - quarterly indicator

#### Output variable (desired target):
- **21 - y (binary)** : has the client subscribed a term deposit? ('yes', 'no')

````python
df.info()
````

We have **41188 entries** and **21 columns**. The column **'y'** is the one we will try to predict.

The numerical columns were correctly identified by pandas and categorical features were loaded with the object dtype. We convert these columns to the categorical dtype.

````python
cat_cols = [column_name for column_name in df.columns if str(
    df[column_name].dtype) == 'object']
df[cat_cols] = df[cat_cols].astype('category')
````

We replace 'unknown' values with ```np.nan```.

````python
df = df.replace('unknown', np.nan)
````

## 1. Exploratory Data Analysis

### 1.1 Univariate analysis

#### 1.1.1 Target variable

````python
df['y'].value_counts().plot(kind='bar', title='Target variable distribution')
plt.xlabel("Term deposit subscribed")
no_print = plt.ylabel("Count")
````

````python
np.sum(df['y'] == 'yes') / df.shape[0]
````

Clients subscribe to a term deposit in only **11.2%** of the cases. This is a significant **class imbalance** in our target variable which we will need to take into consideration in our models.

#### 1.1.2 Categorical features

Let's take a look at our categorical features distributions.

````python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 15))
for i, column in enumerate(cat_cols[:10]): # last column is the target variable
    df[column].value_counts().plot(
        kind='bar', ax=axes[i//2, i % 2], title=column, rot=45)
fig.subplots_adjust(hspace=1)
````

#### 1.1.3 Numerical features

We will now take a look at the numerical columns.

````python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 15))
for i, column in enumerate([x for x in df.columns if x not in cat_cols]):
    df[column].hist(ax=axes[i//2, i % 2], bins=15)
    axes[i//2, i % 2].set_title(column)

fig.subplots_adjust(hspace=.5)
````

We can notice that age distribution is concentrated between 25 and 50. Maybe we can try later to create categories of age instead of taking all possible ages. Duration has some very big outliers so maybe we will need to take care of them.
Concerning pdays, we see that more than 75% of people were never contacted before for a loan so maybe it would be better to create a binary variable which states whether the person was contacted before instead of this representation.

#### 1.1.4 Missing values

````python
df.isnull().sum(axis=0)
````

We will need to handle the missing values in job, marital, education, default, housing and loan columns.

As seen in the previous plots, the 'default', 'loan' and 'marital' histograms have a highly represented value so we will replace the missing values by the most frequent value for these columns.

In the three other categorical columns with missing values (job, education, housing), we don't have a highly represented value so we will keep na for these columns for the moment. A good filling strategy could be sample matching for these columns with an algorithm such as a KNN.

````python
df['marital'] = df['marital'].fillna('married')
df['default'] = df['default'].fillna('no')
df['loan'] = df['loan'].fillna('no')
````

#### 1.2.2 Features vs target

The correlations are very different depending on the categories, we will do further analysis before our first conclusions. We will factorize the variable we want to predict to continue.

````python
df['y'] = pd.factorize(df['y'])[0]
````

````python
fig, axes = plt.subplots(nrows=10, ncols=2, figsize=(15, 40))

for i, col in enumerate(df.columns):
    if col != "y":
        plt.sca(axes[i//2, i % 2])
        if col in cat_cols:
            sns.barplot(x=col, y="y", data=df)
        else:
            suffix = " (10 quantiles)"
            df[col + suffix] = pd.qcut(df[col], 10,
                                       duplicates='drop', precision=0)
            sns.barplot(x=col + suffix, y="y", data=df)
            df = df.drop(col + suffix, axis=1)
        plt.xticks(rotation=45)

fig.subplots_adjust(hspace=1)
````

- Student and retired people seem to be significantly more likely to accept a loan.

- We notice that some months like March, September, October and December have a very high percentage of loan accepted. However, these months are not very represented in the dataset. We can assume that there was a particular target for these months with very interested customers. We will create a column special_month to differentiate customers.
- Days of week seems to be fairly distributed and the influence on the loan accepted looks small.
- Concerning contact, customer contacted by telephone instead of cellular seem to be unappealing.
- We see that the education is very variable concerning acceptance of loans.
- Days of week seems to be fairly distributed and the influence on the loan accepted looks small.
- Concerning contact, contacting customers by telephone instead of cellular seems unappealing.

We compared the result of a previous marketing campaign with 'poutcome'. We see that when it was a success we have a probabilty of loan accepted.

We will now continue with numerical values.

Let's split our dataset to compare numerical values.

````python
dfYes = df[(df['y'] == 1) & (df['age'] < 100)]
dfNo = df[(df['y'] == 0) & (df['age'] < 100)]
sns.distplot(dfYes['age'], label='age repartition for y=no')
plt.show()
sns.distplot(dfNo['age'])
plt.show()
````

Age repartition is very different for people who contracted a loan and not.

````python
dfYes = df[(df['y'] == 1) & (df['previous'] < 1000)]
dfNo = df[(df['y'] == 0) & (df['previous'] < 1000)]
sns.distplot(dfYes['previous'], label='previous contact repartition for y=no')
plt.show()
sns.distplot(dfNo['previous'])
plt.show()
````

We see that people who contracted a loan are likely to have already been contacted.

````python
dfYes = df[(df['y'] == 1) & (df['emp.var.rate'] < 1000)]
dfNo = df[(df['y'] == 0) & (df['emp.var.rate'] < 1000)]
sns.distplot(dfYes['emp.var.rate'],
             label='previous contact repartition for y=no')
plt.show()
sns.distplot(dfNo['emp.var.rate'])
plt.show()
````

````python
dfYes = df[(df['y'] == 1) & (df['cons.price.idx'] < 1000)]
dfNo = df[(df['y'] == 0) & (df['cons.price.idx'] < 1000)]
sns.distplot(dfYes['cons.price.idx'],
             label='previous contact repartition for y=no')
plt.show()
sns.distplot(dfNo['cons.price.idx'])
plt.show()
````

````python
dfYes = df[(df['y'] == 1)]
dfNo = df[(df['y'] == 0)]
sns.distplot(dfYes['euribor3m'], label='previous contact repartition for y=no')
plt.show()
sns.distplot(dfNo['euribor3m'])
plt.show()
````

Employment variation rate, consommation price index and euribor rate seem to have a positive impact on loan contraction. Economic situation is very correlated with loan contraction.

````python
dfYes = df[(df['y'] == 1)]
dfNo = df[(df['y'] == 0)]
sns.distplot(dfYes['cons.conf.idx'],
             label='previous contact repartition for y=no')
plt.show()
sns.distplot(dfNo['cons.conf.idx'])
plt.show()
````

Consumer confidence is a very important parameter for loan.

The other variables only showed slightly differences and we will not proceed the analysis with it.
